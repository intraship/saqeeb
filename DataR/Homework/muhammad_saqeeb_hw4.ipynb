## Computing Dividends

```r
#Function for Daily dividened
caldiv <- function(df) {
    df %>% mutate(
        change = df$Close - lag(df$Close),
        Adj.change = df$Adj.Close - lag(df$Adj.Close),
        diff = change - Adj.change,
        dividend = diff * df$Close,
    )
}
```

### 1. IBM

```r
x <- ibm
ibm <- filter(select(caldiv(x), Date, dividend), ! is.na(dividend))
ibm
```
