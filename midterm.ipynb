## Computing Dividends

```r
#Function for Daily dividened
caldiv <- function(df) {
    df %>% mutate(
        change = df$Close - lag(df$Close),
        Adj.change = df$Adj.Close - lag(df$Adj.Close),
        diff = change - Adj.change> 0.01,
        dividend = round((diff * df$Close),2),
    )
}
```
