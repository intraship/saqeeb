```r
#Problem 2

n = 1:100;
p = 0;
# use for loop to sum each term
for (k in n){
  # calculate each term an add to p
  p = p + ( (-1)^(k+1) / (2*k-1) );
}
# multiply p with 4 to get the final result
pi = 4 * p;



# use vector 1:100 at the place of k
vec_pi = 4 * sum(  (-1)^((1:100)+1) / (2*(1:100)-1) ) 

# print value of pi calculated by both methods
print(paste("Value of pi using for loop: ", pi))
print(paste("Value of pi using vectorization: ", vec_pi))
```
