```python
v1 = matrix(QQ,[[0, 1, 2, 2]]).transpose()
v2 = matrix(QQ,[[2, -3, -2,-1]]).transpose()
v3 = matrix(QQ,[[1,1,3,3]]).transpose()

A= v1.augment(v2).augment(v3)
show(A)
```

```python
def prj(v,w):
    return (v.transpose()*w)[0,0]/(w.transpose()*w)[0,0]*w

w1 = v1
w2 = v2 -prj(v2,w1)
w3 = 9*(v3 - prj(v3,w1) -prj(v3,w2))

#  ------------------------------------------------
# print out the results
x=var('x')
x4=matrix([[x,x,x,x],[x,x,x,x],[x,x,x,x],[x,x,x,x]])
x3=matrix([[x,x,x],[x,x,x],[x,x,x]])

W = w1.augment(w2).augment(w3)
show(x3, x4, A,W)
show (x3, W.transpose(),W.transpose()*A, W.transpose()*W)
S = diagonal_matrix( [1/3, 1/3, sqrt(2)/6 ] )
Qt = S* W.transpose()
Q = Qt.transpose()
R = Qt * A
show(S, Qt, R)
```

# Problem 3: SVD

```python
A = matrix(QQ,[[-3,4],[-2,11],[-10,5],[0,0]])

AtA = A.transpose()*A
show('A=',A, ',  At A =', AtA, ';   50 is an eigenvalue: ', (AtA - 50*identity_matrix(QQ,2)).rref())
print 'Charpoly ', factor(AtA.charpoly())
```

Since $\lambda_1 = 50$, $\lambda_2 = tr(A)-50 = 113+162-50 = 225$

To get $V$, we need the nullspace and column space of $A^t A - \lambda I$ (the special case of a 2x2 matrix):
$$
A^t A - 50 I = \begin{pmatrix} 63 & -84 \\ -84 & 112 \end{pmatrix}
$$
Thus eigenvector for 50 is (84 63) -> ( 4, 3)/5<br>
Eigenvector for 225 must be orthogonal to this (or first col of $(A^t A - 50 I)$ since the matrix is 2x2) -> (3,-4)/5

```python
V = AtA.right_eigenmatrix()[1]*matrix(QQ, [[3,0],[0,4]])/5
show('V =', 1/5,5*V)
```

```python
Sigma_r = diagonal_matrix([sqrt(225),sqrt(50)])
U_r     = A*V*Sigma_r.inverse()
show(A, ' = ', 1/6,6*U_r,5,1/5*Sigma_r, 1/5,5*V.transpose())
```

```python
print 'PseudoInverse'
A_pseudo = V*Sigma_r.inverse()*U_r.transpose()
show( 1/5,5*V,1/30,30*Sigma_r.inverse(), 1/6,6*U_r.transpose(),' = ',1/(5*30*3),5*30*3*A_pseudo)
```

```python
U=(U_r*matrix([[45,0],[0,10]])).augment(identity_matrix(QQ,4))
E1 = matrix(QQ,identity_matrix(QQ,4)); E1[1,0]=-2;E1[2,0]=-2
show(U)
U1=E1*U
show(E1,U1)
E2 = matrix(QQ, identity_matrix(QQ,4)); E2[2,1]=1
U2=E2*U1
show( E2,U2)
```

```python
U = (U_r*matrix([[45,0],[0,10]])).augment(U2[2,2:].transpose())
U = U.augment(vector([0,0,0,1]))
U = U * diagonal_matrix([2/90,1/sqrt(100), 1/sqrt(18),1])
#U.transpose()*U
show( 'U=',1/6,6* U, ',  Sigma=', Sigma_r.augment(matrix([[0,0],[0,0]]) ).transpose(),  ',   V=',1/5,5*V)
```

```python
b = matrix(4,1,[0,-2,1,0])
x_star = A_pseudo*b
b_parallel = A*x_star
b_perp     = b - b_parallel

show('xstar     = ',1/450,450*x_star)
show('bparallel = ', 1/18,18*b_parallel)
show('bperp     = ', 1/18,18*b_perp)
```

Substituting $y = S \tilde{y}$, we get  $y = S e^{\Lambda t} S^{-1} y_0$

# Problem 9
As set of vectors $B = \{v_1, v_2, \dots v_n\}$ is linearly independent iff
$\sum_{i=1}^n { \alpha_i v_i } = 0 \Rightarrow \alpha_i = 0$ for $i=1,2, \dots n$.

The set of vectors $B$ forms a basis for a vector space $V$ iff a) the vectors are linearly independent
and b) span $V$.

Given a matrix $A$ of size $M \times N$, the SVD $A = U \Sigma V^t$, where
* $\Sigma = \begin{pmatrix} \Sigma_r & 0 \\ 0 & 0 \end{pmatrix}$ of size $M \times N$,
where $\Sigma_r$ is a diagonal matrix of size $r \times r$ with non-zero entries in decreasing order on the diagonal. $r = rank(A)$.
* $U$ and $V$ are orthogonal matrices of size $M \times M$ and $N \times N$ respectively.
  * The first $r$ columns of $V$ are a basis for $\mathscr{R}(A)$, the remaining columns
    are  a basis for $\mathscr{N}(A)$
  * the first $r$ columns of $U$ are a basis for $\mathscr{C}(A)$, the remaining columns
    are  a basis for $\mathscr{N}(A^t)$.

# Problem 10
We saw three solution methods for the projection problem $argmin_x || A x - b ||$:
* using the normal equation: solve $A^t A x = A^t b$
* using the QR decomposition: $A = Q R$; solve $R x = Q^t b$
* using the pseudoinverse $A^\dagger$: compute $x = A^\dagger b$.

The vector $b$ decomposes into $b = b_c + b_n$, where $b_c \in \mathscr{C}(A)$ and $b_n \in \mathscr{N}(A^t)$:<br>
$\quad b_c = A x$<br>
$\quad b_n = b - b_c$.
